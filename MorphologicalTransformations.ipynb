```python
from Morph import *
from matplotlib import pyplot as plt
import numpy as np
```

```python
image_path: str = 'Images/fingerPrint1.png'
```

```python
input_image: np.array = load_image(filename=image_path)
```

```python
def show_image(image: np.array):
    plt.figure(figsize=(8,8))
    plt.imshow(image.astype('uint8'), cmap=plt.get_cmap('gray'), vmin=0, vmax=255)
    plt.show()
```

```python
show_image(image=input_image)
```

```python
input_image_gray: np.array = convert_to_gray(input_image=input_image)
```

```python
show_image(image=input_image_gray)
```

```python
input_image_gray = invert_image(
    input_image=input_image_gray)
```

```python
input_image_binary: np.array = convert_to_binary(
    input_image=input_image_gray)
```

```python
show_image(image=input_image_binary)
```

```python
erosion_level: int = 3
```

```python
default_kernel: np.array = np.full(
    shape=(erosion_level, erosion_level),
    fill_value=255)
```

```python
kernel: np.array = default_kernel
```

```python
work_image = opening(
    input_image=input_image_binary,
    kernel=kernel)
```

```python
show_image(image=work_image)
```

```python
work_image = opening(
    input_image=work_image,
    kernel=kernel)
```

```python
show_image(image=work_image)
```

```python
work_image = closing(
    input_image=work_image,
    kernel=kernel)
```

```python
show_image(image=work_image)
```

```python
work_image = erosion(
    input_image=input_image_binary,
    kernel=kernel)
```

```python
show_image(image=work_image)
```

```python
work_image = dilation(
    input_image=work_image,
    kernel=kernel)
```

```python
show_image(image=work_image)
```
